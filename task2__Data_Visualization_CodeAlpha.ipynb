```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_amazon_reviews.csv")
# Preview data
df.head()
```

# Ratings Distribution

This bar chart shows the frequency of each rating (from 1 to 5 stars) given by users.  
It helps us understand how customers generally feel about the product. A high number of 5-star ratings indicates strong satisfaction.

```python
#Visualization 1 - Ratings Distribution

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

sns.countplot(data=df, x='reviews.rating', palette='Blues')

plt.title("Distribution of Product Ratings")
plt.xlabel("Rating (1 to 5 Stars)")
plt.ylabel("Number of Reviews")
plt.tight_layout()
plt.show()
```

## Recommendation Pie Chart

This pie chart shows the percentage of users who recommended the product.  
It gives a clear idea of customer trust and satisfaction — useful for marketing and quality assurance teams.


```python
# Visualization 2 - Recommendation Pie Chart

plt.figure(figsize=(6, 6))

recommend_counts = df['reviews.doRecommend'].value_counts()
recommend_labels = ['Recommended', 'Not Recommended']

plt.pie(recommend_counts,
        labels=recommend_labels,
        autopct='%1.1f%%',
        startangle=140,
        colors=['lightgreen', 'salmon'])

plt.title("Customer Recommendation Distribution")
plt.tight_layout()
plt.show()
```

## Review Length Distribution

This histogram shows how long the average review is in terms of word count.  
It helps us see whether most customers leave brief comments or write detailed feedback.


```python
# Visualization 3 - Review Length Distribution

plt.figure(figsize=(10, 6))
sns.histplot(df['review_length'], bins=30, kde=True, color='purple')

plt.title(" Distribution of Review Lengths")
plt.xlabel("Review Length (Number of Words)")
plt.ylabel("Number of Reviews")
plt.tight_layout()
plt.show()
```

## Review Length by Rating (Boxplot)

This boxplot compares how long reviews are for each star rating.  
Lower ratings tend to have longer reviews, often due to detailed complaints.  
Higher ratings usually have shorter, more positive comments.


```python
# Visualization 4 - Review Length by Rating

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='reviews.rating', y='review_length', palette='Set2')

plt.title(" Review Length vs Rating")
plt.xlabel("Rating (Stars)")
plt.ylabel("Review Length (Words)")
plt.tight_layout()
plt.show()
```

## Dashboard View

This section provides a full dashboard view showing all key visualizations:
- Ratings breakdown
- Customer recommendation share
- Review length distribution
- Comparison of review length by rating

This helps stakeholders quickly understand the overall picture in one glance.


```python
#  Dashboard

fig, axs = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle(' Amazon Kindle Review Dashboard', fontsize=16)

# Chart 1: Ratings Distribution
sns.countplot(data=df, x='reviews.rating', palette='Blues', ax=axs[0, 0])
axs[0, 0].set_title("⭐ Ratings Distribution")
axs[0, 0].set_xlabel("Rating")
axs[0, 0].set_ylabel("Count")

# Chart 2: Recommendation Pie Chart
recommend_counts = df['reviews.doRecommend'].value_counts()
recommend_labels = ['Recommended', 'Not Recommended']
axs[0, 1].pie(recommend_counts, labels=recommend_labels, autopct='%1.1f%%',
              startangle=140, colors=['lightgreen', 'salmon'])
axs[0, 1].set_title(" Recommendation Share")

# Chart 3: Review Length Histogram
sns.histplot(df['review_length'], bins=30, kde=True, color='purple', ax=axs[1, 0])
axs[1, 0].set_title("Review Length Distribution")
axs[1, 0].set_xlabel("Length (words)")
axs[1, 0].set_ylabel("Frequency")

# Chart 4: Boxplot of Review Length by Rating
sns.boxplot(data=df, x='reviews.rating', y='review_length', palette='Set2', ax=axs[1, 1])
axs[1, 1].set_title("Review Length by Rating")
axs[1, 1].set_xlabel("Rating")
axs[1, 1].set_ylabel("Review Length")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Leave space for main title
plt.show()
```
