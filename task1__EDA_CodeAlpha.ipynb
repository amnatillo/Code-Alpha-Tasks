#Data Cleaning

We dropped rows with missing ratings or text and filled missing titles/usernames with default values.

```python
# Drop unnecessary columns
df_clean = df[[
    'name', 'brand', 'categories', 'reviews.rating',
    'reviews.text', 'reviews.title', 'reviews.doRecommend',
    'reviews.numHelpful', 'reviews.username'
]]

# Check for missing values
df_clean.isnull().sum()
```

```python
# Drop rows with missing review text or rating
df_clean = df_clean.dropna(subset=['reviews.text', 'reviews.rating'])

# Fill missing titles/usernames with "Unknown"
df_clean['reviews.title'] = df_clean['reviews.title'].fillna("No Title")
df_clean['reviews.username'] = df_clean['reviews.username'].fillna("Anonymous")

# Confirm shape
df_clean.shape
```

# Explore the Data

```python
# Basic stats
print("Average rating:", df_clean['reviews.rating'].mean())
print("Median helpful votes:", df_clean['reviews.numHelpful'].median())

# Ratings distribution
df_clean['reviews.rating'].value_counts().sort_index()
```

##  Null Value Heatmap

This heatmap shows where missing values are present in the dataset.  
Most missing data is found in the `reviews.id` and `reviews.dateAdded` columns.


```python
# Null Value Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu")
plt.title("Heatmap of Missing Values")
plt.show()
```

## Histogram of Ratings

This histogram displays the distribution of product review ratings (1 to 5 stars).


```python
# Histogram for reviews.rating
plt.figure(figsize=(8, 5))
sns.histplot(df['reviews.rating'], bins=5, kde=False, color='skyblue')
plt.title(" Distribution of Review Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

## Outlier Detection â€“ Helpful Votes

The boxplot identifies reviews with unusually high helpful votes, which might be influential or suspicious.


```python
# Boxplot for Number of Helpful Votes
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='reviews.numHelpful', color='lightcoral')
plt.title("Boxplot of Helpful Votes (Outlier Detection)")
plt.xlabel("Number of Helpful Votes")
plt.show()
```

## Correlation Matrix

This heatmap shows relationships between numeric columns like helpful votes and ratings.


```python
# Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```

## Scatterplot â€“ Helpful Votes vs Ratings

This scatterplot helps detect any relationship between rating scores and how helpful people found the reviews.


```python
# Scatterplot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='reviews.rating', y='reviews.numHelpful', alpha=0.6)
plt.title("Helpful Votes vs Ratings")
plt.xlabel("Review Rating")
plt.ylabel("Number of Helpful Votes")
plt.show()
```

# EDA Dashboard

```python
# EDA Summary Dashboard

fig, axs = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("EDA Dashboard â€“ Amazon Kindle Reviews", fontsize=16)

# 1. Histogram
sns.histplot(df['reviews.rating'], bins=5, kde=False, color='skyblue', ax=axs[0, 0])
axs[0, 0].set_title("â­ Ratings Histogram")
axs[0, 0].set_xlabel("Rating")
axs[0, 0].set_ylabel("Count")

# 2. Boxplot
sns.boxplot(data=df, x='reviews.numHelpful', color='lightcoral', ax=axs[0, 1])
axs[0, 1].set_title("Helpful Votes Boxplot")
axs[0, 1].set_xlabel("Helpful Votes")

# 3. Correlation Heatmap
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, ax=axs[1, 0])
axs[1, 0].set_title("Correlation Matrix")

# 4. Scatterplot
sns.scatterplot(data=df, x='reviews.rating', y='reviews.numHelpful', alpha=0.6, ax=axs[1, 1])
axs[1, 1].set_title("Helpful Votes vs Ratings")
axs[1, 1].set_xlabel("Rating")
axs[1, 1].set_ylabel("Helpful Votes")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
